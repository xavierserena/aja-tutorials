```python
import pandas as pd
import numpy as np
import requests
from prettyprinter import pprint
from nba_api.stats import endpoints
from matplotlib import pyplot as plt

# Here we access the leagueleaders module through endpoints & assign the class to "data"
data = endpoints.alltimeleadersgrids.AllTimeLeadersGrids()
```

```python
df = data.pts_leaders.get_data_frame()
df
```

```python
bars, height = df['PLAYER_NAME'],df['PTS']

xpos = [i*2 for i in range(10)]
fig = plt.figure(figsize=(15,10))
ax = fig.add_axes([0,0,1,1])
ax.set_ylim(25000, 40000)
plt.bar(xpos, height)
plt.xticks(xpos, bars)
plt.show()
```
